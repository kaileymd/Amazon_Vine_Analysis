```python
import pandas as pd
```

```python
df = pd.read_csv("vine_table.csv")
df.head()
```

```python
df.dtypes
```

```python
votes_over_20 = df.loc[df["total_votes"] >= 20]
votes_over_20
```

```python
helpful_votes = votes_over_20.loc[votes_over_20["helpful_votes"] / votes_over_20["total_votes"] >= .5]
helpful_votes
```

```python
#Helpful reviews where a review was paid for
paid_votes = helpful_votes.loc[helpful_votes["vine"] == 'Y']
paid_votes
```

```python
#Helpful reviews where a review was NOT paid for
unpaid_votes = helpful_votes.loc[helpful_votes["vine"] == 'N']
unpaid_votes
```

```python
total_paid_votes = paid_votes["review_id"].count()
total_paid_5_star = paid_votes.loc[paid_votes["star_rating"] == 5]
total_paid_5_star = total_paid_5_star["review_id"].count()
percent_paid_5_star = (total_paid_5_star / total_paid_votes) * 100
```

```python
print(f"Vine program paid for {total_paid_votes} votes")
print(f"Vine program paid for {total_paid_5_star} of their 5 star votes")
print(f"{percent_paid_5_star}% of reviews paid for by Vine were 5 stars")
```

```python
total_unpaid_votes = unpaid_votes["review_id"].count()
total_unpaid_5_star = unpaid_votes.loc[unpaid_votes["star_rating"] == 5]
total_unpaid_5_star = total_unpaid_5_star["review_id"].count()
percent_unpaid_5_star = (total_unpaid_5_star / total_unpaid_votes) * 100
```

```python
print(f"{total_unpaid_votes} votes were not paid for.")
print(f"{total_unpaid_5_star} of 5 star votes were not paid for.")
print(f"{percent_unpaid_5_star}% of unpaid reviews were 5 stars.")
```
